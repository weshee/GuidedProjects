# 1. Inspecting the regression data
This is one of the most popular kind of time series data used for regression. If you can model the value of a company as it changes over time, you can make predictions about where that company will be in the future. This dataset was also originally provided as part of a [public Kaggle competition](https://www.kaggle.com/dgawlik/nyse).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('data/prices.csv', index_col=0, parse_dates=['date'])
data = data.pivot_table(columns='symbol', values='close', index=data.index)
data = data[['AAPL', 'FB', 'NFLX', 'V', 'XOM']]
print(data.head())

# Loop through each column, plot its values over time
fig, ax = plt.subplots()
for column in data.columns:
    data[column].plot(ax=ax, label=column)
ax.legend()
plt.show()
```
